# Data Loading from Drive & Checking

```
data = pd.read_excel('/content/drive/MyDrive/news_share_data.xlsx')
data.shape
```

```
data.head(15)
```

```
df= data.copy()
```

# Processing the Data

```
summary_statistics = df.describe()

# Printing
summary_statistics
```

# Creating the Distribution Plot

```
plt.figure(figsize=(25, 15))

for i, feature in enumerate(summary_statistics, 1):
    plt.subplot(4, 6, i)
    sns.histplot(df[feature], bins=50, kde=True)
    plt.title(f'Distribution of {feature}')
    plt.tight_layout()

plt.show()
```

# Creating Box Plot to See Outlayer

```
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns

plt.figure(figsize=(20, 10))
for i, feature in enumerate(numerical_features, 1):
    plt.subplot(4, 6, i)
    sns.boxplot(df[feature])
    plt.title(f'Box Plot of {feature}')
    plt.tight_layout()

plt.show()
```

# Analyzing Average Shares Over Time by Publication Date

```
df['published_date'] = pd.to_datetime(df['published_date'])
trend_data = df.groupby(df['published_date'].dt.date).mean()

plt.figure(figsize=(15, 10))
plt.plot(trend_data['shares'])

plt.title('Trend of Average Number of Shares')
plt.xlabel('Time Period')
plt.ylabel('Mean Number of Shares')

plt.xticks(rotation=30)
plt.show()
```

# Exploring the Relationship Between Global Subjectivity and Shares [Scatter Plot]

```
plt.figure(figsize=(15, 10))
plt.scatter(df['global_subjectivity'], df['shares'])
plt.title('Relationship Between Global Subjectivity and Shares')
plt.xlabel('Global Subjectivity')
plt.ylabel('Shares')
plt.show()
```

# Exploring the Relationship Between Global Sentiment Polarity and Shares [Scatter Plot]

```
plt.figure(figsize=(15, 10))
plt.scatter(df['global_sentiment_polarity'], df['shares'])
plt.title('Relationship Between Global Sentiment Polarity and Shares')
plt.xlabel('Global Sentiment Polarity')
plt.ylabel('Shares')
plt.show()
```

# Correlation Matrix for Selected Features (Heatmap)

```
selected_features = ['num_hrefs', 'num_imgs', 'num_videos', 'average_token_length','global_subjectivity', 'global_sentiment_polarity', 'shares']
correlation_matrix = df[selected_features].corr()


plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix for Selected Features (Heatmap)')

plt.xticks(rotation=30)
plt.show()
```
